# Demo for Knowing the what, but not the where in Bayesian optimization. Vu Nguyen, Michael Osborne. ICML 2020

# Black-box optimization given the knowledge of the optimum value
\begin{align}
f^* = argmax_{x \in \mathcal{X}} f(x)
\end{align}

# Contact: Vu Nguyen, vu@ieee.org


```python
import sys
sys.path.insert(0,'..')
sys.path.insert(0,'../..')
from bayes_opt.sequentialBO.bo_known_optimum_value import BayesOpt_KnownOptimumValue

import numpy as np
from bayes_opt.visualization import vis_ERM

from bayes_opt.test_functions import functions
import warnings
```

# Transforming the surrogate function go below the optimum value $f*$


```python
myfunction=functions.fourier(sd=0)

func=myfunction.func

func_params={}
func_params['bounds']=myfunction.bounds
func_params['f']=func
func_params['function']=myfunction


x0=[3.1,4.4,8,9]


# create an empty object for BO using vanilla GP ==========================================
acq_func={}
acq_func['name']='ei'
acq_func['surrogate']='gp'
acq_func['dim']=myfunction.input_dim


acq_params={}
acq_params['acq_func']=acq_func

    
gp_params = {'theta':0.04,'noise_delta':1e-8}
bo=BayesOpt_KnownOptimumValue(gp_params,func_params,acq_params)

init_X=np.reshape(x0,(len(x0),1))
init_Y=func(init_X)
bo.init_with_data(init_X=init_X,init_Y=init_Y)


    
# create an empty object for BO using transformed GP =========================================
gp_params = {'lengthscale':0.035,'noise_delta':1e-8}

acq_func={}
acq_func['name']='kov_erm'
acq_func['surrogate']='tgp'

acq_params={}
acq_params['acq_func']=acq_func

bo_tgp=BayesOpt_KnownOptimumValue(gp_params,func_params,acq_params)
bo_tgp.init_with_data(init_X=init_X,init_Y=init_Y)




vis_ERM.plot_1d_Fourier_GP_TGP(bo,bo_tgp,fstar=myfunction.fstar)
```

# Transforming the surrogate function closer to the optimum value $f*$

```python
myfunction=functions.forrester(sd=0)
func=myfunction.func

x0=[0.1,0.46,0.91]

    
# create an empty object for BO using vanilla GP ==========================================
acq_func={}
acq_func['name']='ei'
acq_func['surrogate']='gp'
acq_func['dim']=myfunction.input_dim

func_params={}
func_params['bounds']=myfunction.bounds
func_params['f']=func
func_params['function']=myfunction

acq_params={}
acq_params['acq_func']=acq_func

init_X=np.reshape(x0,(len(x0),1))
init_Y=func(init_X)

bo=BayesOpt_KnownOptimumValue(gp_params,func_params,acq_params)
bo.init_with_data(init_X=init_X,init_Y=init_Y)

    
# create an empty object for BO using transformed GP =========================================
gp_params = {'lengthscale':0.031,'noise_delta':1e-10}

acq_func={}
acq_func['name']='kov_erm'
acq_func['surrogate']='tgp'

acq_params={}
acq_params['acq_func']=acq_func

bo_tgp=BayesOpt_KnownOptimumValue(gp_params,func_params,acq_params,verbose=1)
bo_tgp.init_with_data(init_X=init_X,init_Y=init_Y)




vis_ERM.plot_1d_Forrester_GP_TGP(bo,bo_tgp,fstar=myfunction.fstar)
```

# demonstrating different acquisition functions given known optimum value 

```python
myfunction=functions.fourier(sd=0)
func=myfunction.func

x0=[3.2,4.4,8,9]

gp_params = {'theta':0.033,'noise_delta':1e-8}


# create an empty object for BO using vanilla GP ==========================================
acq_func={}
acq_func['name']='ei'
acq_func['surrogate']='gp'
acq_func['dim']=myfunction.input_dim

func_params={}
func_params['bounds']=myfunction.bounds
func_params['f']=func
func_params['function']=myfunction

acq_params={}
acq_params['acq_func']=acq_func
    
bo=BayesOpt_KnownOptimumValue(gp_params,func_params,acq_params)

init_X=np.reshape(x0,(len(x0),1))
init_Y=func(init_X)
bo.init_with_data(init_X=init_X,init_Y=init_Y)

NN=1*myfunction.input_dim
for index in range(0,NN):
    vis_ERM.plot_acq_bo_1d(bo,fstar=myfunction.fstar)

    
# create an empty object for BO using transformed GP =========================================
acq_func={}
acq_func['name']='kov_erm'
acq_func['surrogate']='tgp'


acq_params={}
acq_params['acq_func']=acq_func

gp_params = {'theta':0.033,'noise_delta':1e-8}

bo_tgp=BayesOpt_KnownOptimumValue(gp_params,func_params,acq_params)
bo_tgp.init_with_data(init_X=init_X,init_Y=init_Y)

NN=1*myfunction.input_dim
for index in range(0,NN):
    vis_ERM.plot_acq_bo_1d_tgp(bo_tgp,fstar=myfunction.fstar)
```

# Running multiple iterations
# using vanilla GP

```python
counter = 0

myfunction=functions.forrester(sd=0)
func=myfunction.func

gp_params = {'lengthscale':0.031,'noise_delta':1e-10}

# create an empty object for BO using vanilla GP
acq_func={}
acq_func['name']='ei'
acq_func['surrogate']='gp'
acq_func['dim']=myfunction.input_dim


func_params={}
func_params['bounds']=myfunction.bounds
func_params['f']=func
func_params['function']=myfunction


acq_params={}
acq_params['acq_func']=acq_func


x0=[0.1,0.46,0.91]


init_X=np.reshape(x0,(len(x0),1))
init_Y=func(init_X)


bo=BayesOpt_KnownOptimumValue(gp_params,func_params,acq_params)

bo.init_with_data(init_X=init_X,init_Y=init_Y)

vis_ERM.plot_1d_Forrester_EI_ERM(bo,fstar=myfunction.fstar)

NN=5*myfunction.input_dim
for index in range(0,NN):
    xt=bo.maximize()
    vis_ERM.plot_1d_Forrester_EI_ERM(bo,fstar=myfunction.fstar)
```

# using Transformed GP and ERM

```python
acq_func={}
acq_func['name']='kov_erm'
acq_func['surrogate']='tgp'


acq_params={}
acq_params['acq_func']=acq_func

gp_params = {'lengthscale':0.031,'noise_delta':1e-10}


bo_tgp=BayesOpt_KnownOptimumValue(gp_params,func_params,acq_params,verbose=1)
bo_tgp.init_with_data(init_X=init_X,init_Y=init_Y)

vis_ERM.plot_1d_tgp_Forrester_EI_ERM(bo_tgp,fstar=myfunction.fstar)


NN=5*myfunction.input_dim
for index in range(0,NN):
    xt=bo_tgp.maximize()
    print(bo_tgp.X_original[-1])

    vis_ERM.plot_1d_tgp_Forrester_EI_ERM(bo_tgp,fstar=myfunction.fstar)
```
